# BUSINESS LOGIC

```python
query = """
    SELECT
        crm_cust.customer_id,
        crm_cust.customer_key,
        crm_cust.first_name,
        crm_cust.last_name,
        crm_cust.marital_status,
        erp_cust_loc.country,
        CASE 
        WHEN crm_cust.gender <> 'n/a' THEN crm_cust.gender
        ELSE coalesce(erp_cust.gender, 'n/a')
        END as gender,
        erp_cust.birth_date,
        crm_cust.create_date
    FROM workspace.silver.crm_customers as crm_cust
    LEFT JOIN workspace.silver.erp_customers as erp_cust
        ON crm_cust.customer_key = erp_cust.customer_key
    LEFT JOIN workspace.silver.erp_customer_location as erp_cust_loc
        ON crm_cust.customer_key = erp_cust_loc.customer_key
    ORDER BY crm_cust.customer_key;
"""

```

```python
df = spark.sql(query)
```

```python
df.limit(10).display()
```

# Writing GOLD Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_customers")
```

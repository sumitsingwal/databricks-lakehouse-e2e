# BUSINESS LOGIC

```python
query = """
    SELECT
        prd.product_id,
        prd.product_key,
        prd.category_id,
        prd.product_name,
        prd_cat.category,
        prd_cat.subcategory,
        prd_cat.maintenance_flag,
        prd.product_cost,
        prd.product_line,
        prd.start_date,
        prd.end_date
    FROM workspace.silver.crm_products as prd
    LEFT JOIN workspace.silver.erp_product_category as prd_cat
        ON prd.category_id = prd_cat.category_id
    ORDER BY prd.product_id
"""
```

```python
df = spark.sql(query)
```

```python
df.limit(10).display()
```

# Writing GOLD Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```

# BUSINESS LOGIC

```python
query = """
    SELECT
        sal.order_number,
        prd.product_key,
        cst.customer_key,
        sal.order_date,
        sal.ship_date,
        sal.due_date,
        sal.sales_amount,
        sal.quantity,
        sal.price
    FROM workspace.silver.crm_sales as sal
    LEFT JOIN workspace.gold.dim_products as prd
        ON sal.product_key = prd.product_key
    LEFT JOIN workspace.gold.dim_customers as cst
        ON sal.customer_id = cst.customer_id;
"""
```

```python
df = spark.sql(query)
```

```python
df.limit(10).display()
```

# Writing GOLD Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
```
